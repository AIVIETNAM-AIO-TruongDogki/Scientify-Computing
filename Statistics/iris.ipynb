```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt
```

# Get general dataset infor

```python
iris = load_iris()

X = iris.data
y = iris.target
feature_names = np.array(iris.feature_names)
target_names = np.array(iris.target_names)

print(X.shape)
print(y.shape)
print(feature_names.tolist())
print(target_names.tolist())
```

```python
class_count = np.bincount(y)    # counting the target (classes)
print(class_count)  

means = np.mean(X, axis=0, keepdims=True)
print(f"Means: {means}")
std = np.std(X, axis=0, keepdims=True)
print(f"Std: {std}")
```

Standardize the X -> std will be 1
Use z (mean + z* std) as a data instead of raw

```python
X_standarded = (X - means) /std
print(X_standarded.shape)
print(X[:5])
print(X_standarded[:5])
```

```python
print(X_standarded.mean(axis=0))
print(X_standarded.std(axis=0))       # std = 1 -> easier for later preprocessing
```

Make a visual compare

```python
# plt.hist(X[:, 2], bins=30)      # bins mean the number of block( devided class)
plt.hist(X[y == 0, 2], bins=10, alpha=0.7, label="setosa")
plt.hist(X[y == 1, 2], bins=10, alpha=0.7, label="versicolor")
plt.hist(X[y == 2, 2], bins=10, alpha=0.7, label="virginica")
plt.title("Histogram of X petal length")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# plt.hist(X_standarded[:,2], bins=30)
plt.hist(X_standarded[y == 0, 2], bins=10, alpha=0.7, label="setosa")
plt.hist(X_standarded[y == 1, 2], bins=10, alpha=0.7, label="versicolor")
plt.hist(X_standarded[y == 2, 2], bins=10, alpha=0.7, label="virginica")
plt.title("Histogram of X petal_legth after normalization")
plt.xlabel("z Value")
plt.ylabel("Frequency")
plt.show()
```
