```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.random.randint(3, 10, (9, 2))
C = np.array([[1], [3]])
print(X)
print(np.linalg.norm((X[:, np.newaxis, :] - C), axis=2))

print((X[:, np.newaxis, :] - C).shape)
```

```python
data = {        # Initialize data
    "Index" : [1, 2, 3, 4, 5, 6, 7, 8, 9],
    "Age" : [18, 20, 22, 30, 34, 40, 60, 66, 70],
    "Expenditure": [80, 90, 85, 50, 64, 60, 30, 40, 25]
}

df = pd.DataFrame(data)

print(df)
```

```python
X = df.to_numpy()[:, 1:]
print(X)
```

```python
np.random.seed(42)
C = X[np.random.choice(X.shape[0], size=3, replace=True)]
print(C)
```

```python
distance = np.linalg.norm(X[:, np.newaxis, :] - C, axis = 2)
print(distance)
```

```python
label = np.argmin(distance,axis = 1)
print(label)
```

```python
for i in range(3):
    C[i] = X[label == i].mean()

print(C)
```
