# Loc vs index

Loc -> only use index (label) <br>
Iloc -> order <br>
while [] can use number of order but index = [10, 20, 30] -> data[0] -> FALSE <br>
then we use "illoc"

```python
print(data_sample.loc[data_sample > 3])
print(data.loc['C'])
print(data.loc['J' : 'S'])
print(data[2 : 4])
```

# Interpolation


```python
data = pd.Series([np.nan, 1, 6, 3, 8, np.nan, 7, np.nan, 2, np.nan],
                name='num_droped')

print(data)
```

```python
data.argmax()                       # -> order      np.int64(3)
data.idxmax()                       # -> index      3
```

```python
data.dropna().reset_index(drop=True)
data.fillna(1.0)
data.interpolate(method='linear', limit_direction='forward')            
```
