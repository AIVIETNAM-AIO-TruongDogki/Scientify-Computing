```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('datasets/StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

# Get the score comparison between male and female students

```python
df_grouped = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

df_grouped.plot(kind='bar', figsize=(8,5))
plt.show()
```

```python
df_long = df.melt(
    id_vars=['parental level of education'],
    value_vars=['math score', 'reading score', 'writing score'],
    var_name='Subject',
    value_name='Score'
)

fig = px.box(
    df_long,
    x='parental level of education',
    y='Score',
    facet_col='Subject'
)

fig
```
