```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../pandas/weather_test.csv')
data.head()
```

```python
sns.jointplot(x='temperature_c',
              y='rainfall_mm',
              data=data[:50],
              hue='weather')

plt.show()
```

```python
sns.stripplot(x='weather',
              y='temperature_c',
              data=data[:50],
              hue='weather')
plt.show()
```

```python
sns.pairplot(data=data[:100],
             diag_kind='kde',
             hue='weather')
```
