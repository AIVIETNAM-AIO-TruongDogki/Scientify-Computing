```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
cancer = load_breast_cancer()

X = pd.DataFrame(cancer.data, columns=cancer.feature_names)
y = pd.Series(cancer.target, name="target")
```

```python
val_size = 0.2      # train_size = 1 - 0.2 - 0.125 = 0.675
test_size = 0.125
random_state = 42
is_shuffle = True

X_train, X_val, y_train, y_val = train_test_split(
    X, y,
    test_size=val_size,
    shuffle=is_shuffle,
    random_state=random_state
)
# tach X -> X_train, X_val
# X_train -> X_train, X_test
X_train, X_test, y_train, y_test = train_test_split(
    X_train, y_train,
    test_size=test_size,
    shuffle=is_shuffle,
    random_state=random_state
)
print(f"Training set size: {len(X_train)}")
print(f"Validation set size: {len(X_val)}")
print(f"Test set size: {len(X_test)}")
```

# Data preprocessing


```python
scaler = StandardScaler()
# Tinh lai mean va chuan hoa lai tu X_train (nguy, do lech chuan)
X_train_scaled = scaler.fit_transform(X_train)

# transform = su dung mean vaa ... phia tren
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

print("First row of training data before and after scaling:")
for col, before, after in zip(X_train.columns, X_train.iloc[0], X_train_scaled[0]):
    print(f"{col:12s}: {before:8.4f} → {after:8.4f}")
```

# KNN classifier implementation

```python
k = 5
knn_classifier = KNeighborsClassifier(n_neighbors=k)
knn_classifier.fit(X_train_scaled, y_train)
```

# Evaluate the Model


```python
y_pred = knn_classifier.predict(X_val_scaled)

accuracy = accuracy_score(y_val, y_pred)
print(f"Accuracy {accuracy:.4f}")


print("\nClassification Report:")
print(classification_report(y_val, y_pred, target_names=cancer.target_names))
```

Make prediction on test set


```python
y_pred = knn_classifier.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy {accuracy:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=cancer.target_names))
```
