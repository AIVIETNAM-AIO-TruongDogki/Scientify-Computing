```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
data = {        # Initialize data
    "Index" : [1, 2, 3, 4, 5, 6, 7, 8, 9],
    "Age" : [18, 20, 22, 30, 34, 40, 60, 66, 70],
    "Expenditure": [80, 90, 85, 50, 64, 60, 30, 40, 25]
}

df = pd.DataFrame(data)

print(df)
```

```python
X1 = df.to_numpy()[:, 1:]
print(X1)
```

```python
np.random.seed(42)

indices = np.random.choice(X1.shape[0], size=3, replace=True)
centroids = X1[indices]

print(centroids)
```

# Implement with numpy
X1(9, 2) -> (9, 1,2) -> (9, 3, 2) 
3: 1 point - k centroids  
cenrtoids (3,2) -> broadcasting -> (1, 3, 2) -> (9, 3, 2)

axis=2 -> norm x + y -> euclidean distance to k centroid

```python
distances = np.linalg.norm(X1[:, np.newaxis, :] - centroids[np.newaxis, :, :], axis=2)
print(distances)
```

# Another way implementation withoud numpy

```python
distances1 = ((X1[:, np.newaxis, :] - centroids) ** 2).sum(axis=2) ** 0.5
print(distances1)
```
