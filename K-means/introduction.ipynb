```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([1.4, 1.0, 1.5, 3.1, 3.7, 4.1])
C = np.array([])
print(X)
print(C)
```

```python
np.random.seed(32)
C = np.random.choice(X, size=(2,))
print(C)
```

```python
def clustering(X, C):
    distance = np.abs(X.reshape(-1, 1) - C)
    # print(distance)
    binary_cls = np.argmin(distance, axis=1)
    # print(binary_cls)
    cls = [[], []]
    cls[0] = np.array(X[binary_cls==0])
    cls[1] = np.array(X[binary_cls==1])
    return cls
    
```

```python
def find_centroid(X, C):
    cls = clustering(X, C)
    C = np.array([cls[0].mean(), cls[1].mean()])
    return C
```

```python
loop = True
step = 0
while(loop):
    if np.all(C == find_centroid(X, C)):
        loop = False
        continue
    C = find_centroid(X, C)
    step += 1
    print(C)
```
