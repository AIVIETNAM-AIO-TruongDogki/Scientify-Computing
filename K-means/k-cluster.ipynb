```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {        # Initialize data
    "Index" : [1, 2, 3, 4, 5, 6, 7, 8, 9],
    "Age" : [18, 20, 22, 30, 34, 40, 60, 66, 70],
    "Expenditure": [80, 90, 85, 50, 64, 60, 30, 40, 25]
}

df = pd.DataFrame(data)

print(df)
```

```python
data = df.to_numpy()[:, 1:]     # remove index column
print(data)
```

```python
X = data
C = np.array([])
```

```python
np.random.seed(10)
def K_means(X, C, k):
    C = X[np.random.choice(len(X), size=k, replace=True)]
    print(C)
    return C
```

```python
def L2_distance(X, C):
    distance = np.array([np.round(np.sqrt(np.sum(np.abs((X - C[i]) ** 2), axis=1)), 2) for i in range (C.shape[0])])
    # X-C[i] = x - xi
    # (x - xi)^2 + (y -yi)^2
    # sqrt()
    # for i -> number of centroids
    return distance.T
    
```

```python
def find_clustering(X, C):
    distance = L2_distance(X, C)
    # print(distance)
    cls = np.array(np.argmin(distance, axis=1))
    cls_array = []
    for i in range (C.shape[0]):
        cls_array.append(np.array(X[cls == i]))
    return cls , cls_array
```

```python
def find_centroid(X, C):
    cls, cls_array = find_clustering(X, C)
    # print(cls)
    # print(cls_array)
    C = np.array([np.mean(cls_array[i], axis=0) for i in range(len(cls_array))])
    return C
```

```python
C = K_means(X, C , 3)
L2_distance(X, C)

loop = True
step = 0

while(loop):
    if np.all(C == find_centroid(X, C)):
        loop = False
        continue
    C = find_centroid(X, C)
    step += 1
    print(f"Step number {step}: ")
    print(C)

print("FINAL RESULT!")
print(C)
print(step)
```
