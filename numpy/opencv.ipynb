# Increasing/Decreasing brightness

```python
img_test = cv2.imread("image_1.jpg", 1)
img_test = img_test.astype(float)

img_test = img_test + 100
img_test = np.clip(img_test, 0, 255)

img_test = img_test.astype(np.uint8)
plt.imshow(img_test)
```

```python
img_test_1 = cv2.imread("image_1.jpg")
img_test_1 = img_test_1.astype(float)

img_test_1 -= 100
img_test_1 = np.where(img_test_1 < 0, 0, img_test_1)
img_test_1 = np.where(img_test_1 > 255, 255, img_test_1)

img_test_1 = img_test_1.astype(np.uint8)
plt.imshow(img_test_1)
```

```python
bg = cv2.imread("background.jpg", 1)
bg = cv2.resize(bg, (1920, 1080))                       # (with, height)  -> CONTRAST to numpy

pt = cv2.imread("portrait.jpg", 1)
pt = cv2.resize(pt, (1920, 1080))


diff = cv2.absdiff(bg, pt)
_, diff_binary = cv2.threshold(diff.astype(np.uint8), 15,
                               255, cv2.THRESH_BINARY)
cv2.imwrite('output.png', diff_binary)
```

```python
new_bg = cv2.imread("shibameme.jpg", 1)
new_bg = cv2.resize(new_bg, (1920, 1080))

new_pt = np.where(diff_binary==0, new_bg, pt)

plt.imshow(new_pt[:, :, [2, 1, 0]])
cv2.imwrite("new_pt.png", new_pt)
```
