```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
x = np.array([1, 2, 3, 4 ,5])
y = np.array([3, 2, 1, 5, 3])

start = time.time()
print(x + y)                        # normal array addition
end = time.time()  

start_1 = time.time()
print(np.add(x, y))                 # np.add -> FASTER
end_1 = time.time()


print(end - start)
print(end_1 - start_1)
```

```python
print(np.subtract(x, y))
print(np.multiply(x , y))
print(np.divide(x, y))
print(x % y)
print(x // y)
```

# Matrix x Vector

```python
X = np.array([[1, 2],
              [3, 4]])

v = np.array([2, 3])            # 1-dimensional -> special case (VECTOR)

print(X.dot(v))                 # 2x2 . 2x1

print(v.dot(X))                 # 1x2 . 2x2 
```

# Matrix x Matrix

```python
v_reshape = v.reshape(1, 2)

print(X.dot(v_reshape.T))
print(v_reshape.dot(X))

print(np.dot(X, v_reshape.T))
print(X@v_reshape.T)
```

```python
start = time.time()
print(v@X)
end = time.time()

start_1 = time.time()
print(np.dot(v, X))
end_1 = time.time()

print(end - start)
print(end_1 - start_1)
```

# Vector operations


```python
color_image = cv2.imread("shibameme.jpg")
color_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2RGB)

print(color_image.shape)

vector = np.array([0.072, 0.715, 0.213])

grayscale_image = np.dot(color_image,vector)                # the last dimensional * vector -> reduce 1 dimen

print(grayscale_image.shape)
cv2.imwrite("output.png", grayscale_image)
```

```python
matrix1 = np.arange(9).reshape(3, 3)

matrix2 = np.ones((3, 3))

print(matrix1)
print(matrix1@matrix2)
```

```python
np.random.seed(42)
matrix = np.random.randint(1, 10, (3, 3))
print(matrix)

matrix = np.argmax(matrix, axis=0)
print(matrix)
```

```python
np.random.seed(42)
matrix = np.random.randint(1, 10, (3, 3))
print(matrix)
_, result = cv2.threshold(matrix.astype(np.uint8), 4,               # <= 4 -> max
                        255, cv2.THRESH_BINARY)
print(result)
```

```python
vector1 = np.array([5, 3, 2, 7])
vector2 = np.array([2, 9, 4, 1])

cosine = (np.dot(vector1, vector2)) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))

print(cosine)
```
